```python
import matplotlib.pyplot as plt
import csv

import numpy as np

# Adjust the path to where your CSV file is saved
csv_file_path = 'timing_result.csv'

# Initialize lists to store the data
m_size = []
brute_force_alg = []
recursive_alg = []
strassen_alg = []

# Open and read the CSV file
with open(csv_file_path, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        m_size.append(int(row[0]))
        brute_force_alg.append(float(row[1]))
        recursive_alg.append(float(row[2]))
        strassen_alg.append(float(row[3]))

# plt.figure(figsize=(10, 6))
plt.plot(m_size, brute_force_alg, marker='o', label='Brute Force')
plt.plot(m_size, recursive_alg, marker='x', label='Recursive')
plt.plot(m_size, strassen_alg, marker='s', label='Strassen')

plt.title('Function Execution Time over different sizes')
plt.xlabel('Matrice Size')
plt.ylabel('Time (milliseconds)')
plt.legend()
plt.grid(True)
plt.show()
```
