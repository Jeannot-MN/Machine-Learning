```python
import matplotlib.pyplot as plt
import numpy as np

x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

plt.scatter(x_data, y_data)
plt.show()
```

```python
plt.scatter(x_data, y_data, c="red")
plt.show()
```

```python
plt.scatter(x_data, y_data, c="#f5f500")
plt.show()
```

```python
plt.scatter(x_data, y_data, c="#f5f500", marker="*")
plt.show()
```

```python
plt.scatter(x_data, y_data, c="#f5f500", marker="*", s=150)
plt.show()
```

```python
plt.scatter(x_data, y_data, c="#f5f500", marker="*", alpha=0.3)
plt.show()
```

```python
years = np.arange(15, 26)
weight = years * 2 - 5

plt.plot(years, weight, c= 'r', lw=2, linestyle="--")
plt.show()
```

```python
plt.plot(years, weight, 'r--', lw=2)
plt.show()
```

```python
ages = np.random.normal(20, 1.5, 1000)

plt.hist(ages)
plt.show()
```

```python
years = np.arange(2014, 2025)
income = np.random.randint(50,110,years.size)

plt.plot(years, income)

plt.title("Your yearly income in USD", fontsize=20)

plt.xlabel("Year")
plt.ylabel("Income")

ticks = [i for i in range(50,120,5)]
plt.yticks(ticks, [f"${x}k\$" for x in ticks])


plt.show()
```

```python
income_1 = np.random.randint(50,80,10)
income_2 = np.random.randint(50,80,10)
income_3 = np.random.randint(50,80,10)

plt.plot(income_1, label="JMN")
plt.plot(income_2, label="Nou")
plt.plot(income_3, label="Bou")

plt.legend(loc="lower right")
plt.show()
```

```python
x = np.arange(90)

fig, axs = plt.subplots(2,2)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sin wave")
axs[1,0].plot(x, np.cos(x))
axs[1,0].set_title("Cosine wave")
axs[0,1].plot(x, np.exp(x))
axs[0,1].set_title("Exp")
axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("Log")

fig.suptitle("4 figs")
plt.tight_layout()
plt.show()
```

```python
ax = plt.axes(projection="3d")

x, y = np.arange(45), np.arange(45)
z= np.sin(x) - np.cos(y)

ax.plot(x,y,z)
plt.show()
```

```python
%matplotlib notebook
ax = plt.axes(projection="3d")
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X,Y,Z, cmap="Spectral")
plt.show()
```
