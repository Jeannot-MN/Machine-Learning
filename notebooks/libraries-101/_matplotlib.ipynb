```python
import matplotlib.pyplot as plt
import numpy as np

x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

plt.scatter(x_data, y_data)
plt.show()
```

```python
plt.scatter(x_data, y_data, c="red")
plt.show()
```

```python
plt.scatter(x_data, y_data, c="#f5f500")
plt.show()
```

```python
plt.scatter(x_data, y_data, c="#f5f500", marker="*")
plt.show()
```

```python
plt.scatter(x_data, y_data, c="#f5f500", marker="*", s=150)
plt.show()
```

```python
plt.scatter(x_data, y_data, c="#f5f500", marker="*", alpha=0.3)
plt.show()
```

```python
years = np.arange(15, 26)
weight = years * 2 - 5

plt.plot(years, weight, c= 'r', lw=2, linestyle="--")
plt.show()
```

```python
plt.plot(years, weight, 'r--', lw=2)
plt.show()
```

```python
x = ["C++","C#","Python","Java","Go"]
votes = [20, 50, 140, 1, 45]

plt.bar(x, votes)
plt.show()
```

```python
plt.bar(x, votes, color='r', align='edge', width=0.4)
plt.show()
```

```python
ages = np.random.normal(20, 1.5, 1000)

plt.hist(ages)
plt.show()
```

```python
plt.hist(ages, bins=[ages.min(), 18, 21, ages.max()])
plt.show()
```

```python
x = ["C++","C#","Python","Java","Go"]
votes= [20, 50, 140, 1, 45]

plt.pie(votes, labels=x)
plt.show()
```

```python
explodes = [0,0.2,0,0, 0]

plt.pie(votes, labels=x, explode=explodes)
plt.show()
```

```python
plt.pie(votes, labels=x, autopct="%.2f%%")
plt.show()
```
